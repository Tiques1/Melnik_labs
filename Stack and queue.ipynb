# 1.	Реализовать Стек с помощью списка. Должны поддерживаться следующие операции работы со стеком: Push(elem), Pop(), Top(), isEmpty(), Print().

```python
class Stack:
    def __init__(self):
        self.__stack = []
        self.__count = 0 
    
    def push(self, x): # Поместить элемент
        self.__stack.append(x)
        self.__count+=1

    def pop(self): # Достать элемент
        if self.isEmpty():
            print('Stack is empty')
        else:
            self.__count-=1
            return self.__stack.pop()

    def top(self): # Посмотреть вершину
        if self.isEmpty():
            print('Stack is empty')
        else:
            elem = self.pop()
            self.push(elem)
            return elem

    def isEmpty(self): # Проверить на пустоту
        return self.__count == 0
    
    def stack(self):
        temp = []
        for i in range(self.__count):
            temp.append(self.__stack.pop())
        for j in range(self.__count - 1, -1, -1):
            self.__stack.append(temp[j])
        return temp
        
    
stack1 = Stack()
print('Stack:', stack1.stack())
stack1.push(5)
stack1.push('cat')
print('Stack:', stack1.stack())
print('Top:', stack1.top())
print('Pop:', stack1.pop())
print('Stack:', stack1.stack())
print('Empty:', stack1.isEmpty())
```

Для операции push после единицы через запятую указывается значение помещаемого элемента (это может быть число либо слово). Результат выполнения каждой операции выводится на экран.

### Пример: 3 4 1,56 1,7 1,cat 2 5 4


```python
import time

def time_decorator(foo):
    def function(*args):
        start_time = time.perf_counter_ns()
        result = foo(*args)
        end_time = time.perf_counter_ns()
        return end_time - start_time
    return function
```

```python
# Операции со стеком из файла
def file(path, stack=Stack()): # Путь к файлу(string), stack(object class Stack)
    with open(path, 'r') as f:
        for i in f:
            print(executor(i.split(), stack))
        
@time_decorator
def executor(commands, stack):
    for command in commands:
        i = command.split(',')
        if int(i[0]) == 1:
            stack.push(i[1])
            print('Push:', i[1])
        elif int(i[0]) == 2:
            print('Pop:', stack.pop())
        elif int(i[0]) == 3:
            print('Top:', stack.top())
        elif int(i[0]) == 4:
            print('Empty:', stack.isEmpty())
        elif int(i[0]) == 5:
            print('Stack:', stack.stack())
            
stack = Stack()
file('./input.txt', stack)
```

### После реализации алгоритма вычисления постфиксной записи, необходимо произвести расчет оценки сложности алгоритма.


```python
def postfix(string, stack): # Сложность = n(6 + 14 + 6) = 26n = O(n)
    for i in string: # n
        if i in '0123456789': # 1
            stack.push(int(i))
        elif i == '+': # 1
            stack.push(stack.pop() + stack.pop()) 
        elif i == '-': # 1
            stack.push(stack.pop() - stack.pop()) 
        elif i == '*': # 1
            stack.push(stack.pop() * stack.pop())
        elif i == '/': # 1
            stack.push(stack.pop() / stack.pop())
        elif i == '^': # 1
            stack.push(stack.pop() ** stack.pop()) # 2 * 5 + 1 + 1 * 3
    return stack.pop() # 1 + 1 * 5
            
stack = Stack()
print(postfix('235+5*^', stack)) # ((3+5)*5)^2
```

# 1. Реализовать Очередь с помощью списка и с помощью стандартного класса Queue. Должны поддерживаться следующие операции работы с очередью: вставка/удаление элемента, проверка на пустоту, печать, вывод первого элемента.

```python
class MyQueue:
    def __init__(self):
        self.__queue = []
        self.__count = 0

    # Вставка элемента
    def put(self, element):
            self.__queue.append(element)
            self.__count += 1

    # Удаление элемента
    def get(self):  
        if self.empty():
            print('Очередь пуста')
        else:
            self.__count -= 1
            return self.__queue.pop(0)

    # Проверка на пустоту
    def empty(self):
        return self.__count == 0

    # Печать очереди
    def queue(self):
        temp = []
        for _ in range(self.__count):
            temp.append(self.get())
        for j in temp:
            self.put(j)
        return temp
            
    # Вывод первого элемента очереди
    def first_elem(self):
        if self.empty():
            print('Очередь пуста')
        else:
            elem = self.queue()[0]
            return elem


q = MyQueue()
q.put(1)
q.put(2)
q.put(3)
print('Queue', q.queue())
print('Первый элемент:', q.first_elem())
print('Извлечен:', q.get())
print('Queue', q.queue())
```
